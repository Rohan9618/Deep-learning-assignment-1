```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score

# Initialize a dictionary to store all metrics
metrics = {
    'Model': [],
    'Accuracy': [],
    'FNR (False Negative Rate)': [],
    'FPR (False Positive Rate)': [],
    'Precision': [],
    'Recall': [],
    'F1-Score': []
}

# Calculate metrics for each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    
    # Extract values from confusion matrix
    TN, FP, FN, TP = cm.ravel()
    
    # Calculate the metrics
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    fnr = FN / (FN + TP)  # False Negative Rate
    fpr = FP / (FP + TN)  # False Positive Rate
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)
    
    # Store the results
    metrics['Model'].append(name)
    metrics['Accuracy'].append(accuracy)
    metrics['FNR (False Negative Rate)'].append(fnr)
    metrics['FPR (False Positive Rate)'].append(fpr)
    metrics['Precision'].append(precision)
    metrics['Recall'].append(recall)
    metrics['F1-Score'].append(f1_score)

# Convert metrics dictionary to a pandas DataFrame for easy viewing
metrics_df = pd.DataFrame(metrics)

# Display the DataFrame
print(metrics_df)
```
